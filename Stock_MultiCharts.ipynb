```
import yfinance as yf
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
def MultiCharts(tickers):
  for ticker in tickers:
    ticker_obj = yf.Ticker(ticker)
    mo_info = ticker_obj.history(period="3mo")
    mo_prices = mo_info[['Open', 'High', 'Low', 'Close']]
    current_price = str(round(mo_prices['Close'][-1], 2))
    mo_high = str(round(mo_prices['High'].max(),2))
    mo_low = str(round(mo_prices['Low'].min(),2))

    wk_info = ticker_obj.history(period="2wk")
    wk_prices = wk_info[['Open', 'High', 'Low', 'Close']]

    fig, ax =plt.subplots(1,2, figsize=[25,5])
    plt.title(ticker+"   CURRENT:   "+current_price+"   3moHIGH:     "+mo_high+"    3moLOW:    "+mo_low)
    sns.lineplot(data=wk_prices, ax=ax[0])
    sns.lineplot(x='Date', y='Close',data=mo_prices, ax=ax[1])
    fig.show()
```

```
my_tickers = ["ETSY", "OPEN", "MILE", "PLTR" ,"ABNB", "SNOW", "AFRM", "BYND", "DASH", "UBER", "NFLX", "PYPL", "PDD", "SHOP", "SNAP", "SPCE", "SPOT", "SQ", "TSLA", "NIO"]
bio_tickers = ["NTRA", "FLGT", "BEAM", "EDIT", "CRSP", "NTLA", "BNTX", "MRNA"]
index_tickers = ["SPY", "SQQQ"]
```

```
MultiCharts(bio_tickers)
```
